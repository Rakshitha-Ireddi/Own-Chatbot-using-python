# MY OWN CHATBOT USING PYTHON


```python
#code by rakshitha ireddi

import random

greetings = ["hi", "hello", "hey", "hi there", "hello there"]
farewells = ["goodbye", "bye", "see you later", "take care", "have a nice day"]
how_are_you = ["how are you", "how's it going", "how are you doing"]
bot_name = ["what's your name", "do you have a name", "who are you", "name"]
technology_questions = ["tell me about AI", "what is machine learning", "explain blockchain", "what is Python", "what is a computer virus"]
more_technology_questions = [
    "Explain cloud computing",
    "What is virtual reality (VR)?",
    "Tell me about the Internet of Things (IoT)",
    "What is a smartphone?",
    "How does 5G technology work?",
    "What are the advantages of open-source software?",
    "Explain the concept of big data",
    "What is a computer algorithm?",
    "Tell me about cybersecurity",
    "What is a data breach?",
    "How does encryption work?",
    "What is machine vision?",
    "Explain the difference between software and hardware",
    "What is the role of a web browser?",
    "Tell me about artificial neural networks",
    "What is a database management system (DBMS)?",
    "What is the significance of user experience (UX) in design?",
    "Explain the concept of agile software development",
    "What are the basic principles of user interface (UI) design?",
    "Tell me about quantum computing"
]

tech_recommendations = ["recommend a programming language", "suggest a good tech book", "what tech podcast should I listen to", "recommend a coding platform"]
business_hours = ["when are you available", "what are your business hours", "hours of operation", "working hours"]
unknown = ["I don't understand that", "Could you please rephrase that", "Sorry, I'm not sure what you mean"]

print("*******************************************************\n")

while True:
    user_input = input('User said - ')

    if user_input.lower() in greetings:
        bot_responses = ["Hello!", "Hi there!", "Greetings!"]
        print('Bot said - ' + random.choice(bot_responses) + '\n')

    elif user_input.lower() in farewells:
        bot_responses = ["Goodbye!", "See you later!", "Take care!"]
        print('Bot said - ' + random.choice(bot_responses) + '\n')

    elif user_input.lower() in how_are_you:
        bot_responses = ["I'm just a computer program, but I'm here to help!", "I'm doing well, thank you!", "I'm here to assist you!"]
        print('Bot said - ' + random.choice(bot_responses) + '\n')

    elif user_input.lower() in bot_name:
        bot_responses = ["You can call me TechBot.", "I go by TechBot.", "I'm TechBot, your tech assistant."]
        print('Bot said - ' + random.choice(bot_responses) + '\n')

    elif any(keyword in user_input.lower() for keyword in technology_questions + more_technology_questions):
        tech_responses = ["Artificial Intelligence (AI) is the simulation of human intelligence in machines.",
                          "Machine Learning is a subset of AI that allows systems to learn from data and make predictions.",
                          "Blockchain is a distributed ledger technology used for secure and transparent transactions.",
                          "Python is a versatile and widely-used programming language.",
                          "A computer virus is a malicious program that can replicate and harm computer systems."]
        print('Bot said - ' + random.choice(tech_responses) + '\n')

    elif any(keyword in user_input.lower() for keyword in tech_recommendations):
        bot_responses = ["I recommend learning Python for its versatility and popularity.",
                         "Consider reading 'Clean Code' by Robert C. Martin for software development best practices.",
                         "You might enjoy the 'Coding Blocks' podcast for tech discussions and tips.",
                         "Check out platforms like GitHub and LeetCode for coding practice."]
        print('Bot said - ' + random.choice(bot_responses) + '\n')

    elif any(keyword in user_input.lower() for keyword in business_hours):
        bot_responses = ["I'm available 24/7 to assist you."]
        print('Bot said - ' + random.choice(bot_responses) + '\n')

    else:
        print('Bot said - ' + random.choice(unknown) + '\n')
```
